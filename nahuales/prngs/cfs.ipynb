# Random walker

```python
import random
import matplotlib.pyplot as plt
import numpy as np

fieldSize = 10
walkers = 4
steps = 20
positions = [[] for w in range(walkers)]
 
# creating the random points
for w in range(walkers):
    for t in range(steps+1):
        positions[w].append(random.randint(0, fieldSize))
    
# plotting down the graph of the random walk in 1D
for w in range(walkers):
    plt.plot(positions[w])

plt.grid(True)
plt.grid(color = '0.5', linestyle = '--', linewidth = 1)
plt.xticks(np.arange(0, steps+1, step=1))
plt.yticks(np.arange(0, fieldSize+1, step=1))

fig = plt.gcf()
fig.set_size_inches(10, 5)
fig.savefig('img/randomwalker.png', dpi=100)

plt.show()
```
